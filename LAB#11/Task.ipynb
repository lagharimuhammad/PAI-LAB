```python
import cv2
import numpy as np
```

```python
# Task 1
image = cv2.imread('image1.jpg')

cv2.imshow('Original Image', image)
cv2.waitKey(0)

gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2.imshow('Grayscale Image', gray_image)
cv2.waitKey(0)

resized_image = cv2.resize(image, (400, 300))
cv2.imshow('Resized Image', resized_image)
cv2.waitKey(0)

cv2.imwrite('grayscale_output.jpg', gray_image)
cv2.imwrite('resized_output.jpg', resized_image)

cv2.destroyAllWindows()
```

```python
# Task 2
blank_image = np.zeros((500, 700, 3), dtype=np.uint8)
blank_image[:] = (255, 255, 255)

cv2.rectangle(blank_image, (50, 50), (250, 200), (0, 255, 0), 3)

cv2.rectangle(blank_image, (300, 100), (500, 250), (255, 0, 0), -1)

cv2.circle(blank_image, (150, 350), 80, (0, 0, 255), 5)

cv2.circle(blank_image, (400, 370), 60, (255, 165, 0), -1)

cv2.rectangle(blank_image, (520, 50), (650, 150), (128, 0, 128), 2)
cv2.circle(blank_image, (585, 300), 50, (0, 255, 255), 4)

cv2.imshow('Blank Image with Shapes', blank_image)
cv2.waitKey(0)

cv2.imwrite('shapes_output.jpg', blank_image)

cv2.destroyAllWindows()
```

```python
# Task 3
image = cv2.imread('image1.jpg')

blurred_image = cv2.GaussianBlur(image, (15, 15), 0)

cv2.imshow('Original Image', image)
cv2.imshow('Blurred Image', blurred_image)
cv2.waitKey(0)

height, width = image.shape[:2]
x_start, y_start = 100, 50
x_end, y_end = 400, 300

cropped_image = image[y_start:y_end, x_start:x_end]

cv2.imshow('Cropped Image', cropped_image)
cv2.waitKey(0)

cv2.imwrite('blurred_output.jpg', blurred_image)
cv2.imwrite('cropped_output.jpg', cropped_image)

cv2.destroyAllWindows()
```

```python
# Task 4
image = cv2.imread('image1.jpg')

font = cv2.FONT_HERSHEY_SIMPLEX
text = 'OpenCV Text'
position = (50, 100)
font_scale = 2
color = (0, 255, 0)
thickness = 3

image_with_text = image.copy()
cv2.putText(image_with_text, text, position, font, font_scale, color, thickness, cv2.LINE_AA)

cv2.imshow('Image with Text', image_with_text)
cv2.waitKey(0)

blank_image = np.zeros((400, 800, 3), dtype=np.uint8)
blank_image[:] = (255, 255, 255)

cv2.putText(blank_image, 'Hello World!', (50, 100), cv2.FONT_HERSHEY_DUPLEX, 2, (255, 0, 0), 3, cv2.LINE_AA)
cv2.putText(blank_image, 'OpenCV Tutorial', (50, 200), cv2.FONT_HERSHEY_TRIPLEX, 1.5, (0, 255, 0), 2, cv2.LINE_AA)
cv2.putText(blank_image, 'Image Processing', (50, 300), cv2.FONT_HERSHEY_COMPLEX, 1.2, (0, 0, 255), 2, cv2.LINE_AA)

cv2.imshow('Blank Image with Text', blank_image)
cv2.waitKey(0)

cv2.imwrite('image_with_text.jpg', image_with_text)
cv2.imwrite('blank_with_text.jpg', blank_image)

cv2.destroyAllWindows()
```

```python
# Task 5
gray_image = cv2.imread('image1.jpg', cv2.IMREAD_GRAYSCALE)

cv2.imshow('Original Grayscale', gray_image)
cv2.waitKey(0)

ret, binary_image = cv2.threshold(gray_image, 127, 255, cv2.THRESH_BINARY)

cv2.imshow('Binary Thresholded Image', binary_image)
cv2.waitKey(0)

height, width = binary_image.shape
center = (width // 2, height // 2)

rotation_matrix = cv2.getRotationMatrix2D(center, 60, 1.0)

rotated_image = cv2.warpAffine(binary_image, rotation_matrix, (width, height))

cv2.imshow('Rotated Image (60 degrees)', rotated_image)
cv2.waitKey(0)

cv2.imwrite('binary_threshold.jpg', binary_image)
cv2.imwrite('rotated_60.jpg', rotated_image)

cv2.destroyAllWindows()
```

```python
# Task 6
image1 = cv2.imread('image1.jpg')
image2 = cv2.imread('image2.jpg')

if image1.shape != image2.shape:
    image2 = cv2.resize(image2, (image1.shape[1], image1.shape[0]))

alpha = 0.5
beta = 0.5
gamma = 0

blended_image = cv2.addWeighted(image1, alpha, image2, beta, gamma)

cv2.imshow('Image 1', image1)
cv2.imshow('Image 2', image2)
cv2.imshow('Blended Image', blended_image)
cv2.waitKey(0)

gray_blended = cv2.cvtColor(blended_image, cv2.COLOR_BGR2GRAY)

cv2.imshow('Grayscale Blended', gray_blended)
cv2.waitKey(0)

equalized_image = cv2.equalizeHist(gray_blended)

cv2.imshow('Histogram Equalized Image', equalized_image)
cv2.waitKey(0)

cv2.imwrite('blended_output.jpg', blended_image)
cv2.imwrite('gray_blended_output.jpg', gray_blended)
cv2.imwrite('equalized_output.jpg', equalized_image)

cv2.destroyAllWindows()
```
